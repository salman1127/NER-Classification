# Loading Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv('pnr.train.txt',sep='\s',header=None)
test=pd.read_csv('pnr.test.txt',sep='\s',header=None,error_bad_lines=False)
```

# Pre Processing

```python
train.drop(2,axis=1,inplace=True)
test.drop(2,axis=1,inplace=True)
```

```python
train=train[train[0]!='.']
test=test[test[0]!='.']
```

```python
train=train[train[0]!='-DOCSTART-']
test=test[test[0]!='-DOCSTART-']
```

```python
train[0]=train[0].str.lower()
train.head()
```

```python
test[0]=test[0].str.lower()
test.head()
```

```python
train.shape,test.shape
```

```python
temp=pd.concat([train,test])
```

```python
temp.shape
```

```python
temp.head()
```

```python
temp[0]=temp[0].astype(str)
```

# Label Encoding

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
le.fit(temp[0])
```

```python
test[0]=test[0].astype(str)
train[0]=train[0].astype(str)
```

```python
test['enco_words']=le.transform(test[0])
```

```python
train['enco_words']=le.transform(train[0])
```

```python
le.fit(temp[1])
```

```python
test['enco_POS']=le.transform(test[1])

train['enco_POS']=le.transform(train[1])
```

```python
test.head()
```

```python
train[3]=train[3].astype(str)
test[3]=test[3].astype(str)
```

```python
train[3].replace('None',0,inplace=True)
train[3].replace('I',1,inplace=True)
train[3].replace('O',0,inplace=True)
```

```python
test[3].replace('None',0,inplace=True)
test[3].replace('I',1,inplace=True)
test[3].replace('O',0,inplace=True)
```

```python
sns.countplot(x=3,data=train)
```

```python
sns.countplot(x=3,data=test)
```

## Applying Classifier

```python
import xgboost as xgb
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
```

```python
train_X=train[['enco_words','enco_POS']]
train_y=train[[3]]

test_X=test[['enco_words','enco_POS']]
test_y=test[[3]]
```

```python
gbm = xgb.XGBClassifier(max_depth=40, n_estimators=300, learning_rate=0.05).fit(train_X, train_y)
predictions = gbm.predict(test_X)
```

```python
print(classification_report(test_y, predictions))
```

```python
f1_score(test_y, predictions, average='weighted')
```
